```python
import cv2
import numpy
import os
import keras
import numpy as np
import pandas as pd
from keras.models import Sequential 
from keras.layers import Dense, Dropout, Flatten 
from keras.layers import Conv2D, MaxPooling2D
import tensorflow as tf
from keras.models import load_model
from sklearn.model_selection import train_test_split
```

```python
{1: "Drowsy", 0: "Alert"}
```

```python
drowsy = os.listdir("../Dataset/Eye Dataset/Drowsy Resized")
alert = os.listdir("../Dataset/Eye Dataset/Alert Resized")
```

```python
df_train = pd.DataFrame(columns=["Input", "Output"])
```

```python
for i in alert:
    frame = cv2.imread("../Dataset/Eye Dataset/Alert Resized/" + i)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    print(frame.shape)
    
    dict_ = {"Input": [frame], "Output": [0]}
    df_train = pd.concat([df_train, pd.DataFrame(data=dict_)], axis=0).reset_index(drop=True)
```

```python
model = Sequential([
    Conv2D(20, kernel_size=3, activation='relu', input_shape=(100, 250, 1)),
    Conv2D(10, kernel_size=3, activation='relu'),
    Conv2D(5, kernel_size=3, activation='relu'),
    Conv2D(3, kernel_size=3, activation='relu'),
    Flatten(),
    Dense(20, activation='relu'),
    Dense(10, activation='relu'),
    Dense(5, activation='relu'),
    Dense(2, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
X_train, X_cv, y_train, y_cv = train_test_split(df_train["Input"], df_train["Output"], train_size=0.7)
```

```python
X_train_ = []
X_cv_ = []
y_train_ = []
y_cv_ = []

for i in X_train:
    X_train_.append(i)

for i in X_cv:
    X_cv_.append(i)
    
for i in y_train:
    y_train_.append(i)
    
for i in y_cv:
    y_cv_.append(i)
    
```

```python
X_train_ = np.array(X_train_)
X_cv_ = np.array(X_cv_)
y_train_ = np.array(y_train_)
y_cv_ = np.array(y_cv_)
```

```python
X_train = tf.convert_to_tensor(X_train_)
X_cv = tf.convert_to_tensor(X_cv_)
```

```python
y_train = y_train_.astype("float32")
y_cv = y_cv_.astype("float32")
```

```python
model.fit(X_train, y_train, validation_data=(X_cv, y_cv), epochs=100, batch_size=20)
```

```python
model.save("Eye Drowsiness Detection")
```
