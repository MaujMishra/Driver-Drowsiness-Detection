```python
frames_lst = os.listdir("../Dataset/" + folder + "/Frames")
```

```python
with mp_face_mesh.FaceMesh(max_num_faces=1, refine_landmarks=True, min_detection_confidence=0.6, min_tracking_confidence=0.6) as face_mesh:
    
    for frame_name in frames_lst:
        #print(frame_name)
        #frame_name = "600.jpg"
        
        index = frames_lst.index(frame_name)
        print(str(index + 1) + "/" + str(len(frames_lst)))
        
        
        frame_path = "../Dataset/" + folder + "/Frames/" + frame_name
        
        frame_1 = cv2.imread(frame_path)

        frame = cv2.cvtColor(frame_1, cv2.COLOR_BGR2RGB)

        results = face_mesh.process(frame)

        img_h, img_w = frame.shape[:2]
        
        try:
            mesh_points=np.array([np.multiply([p.x, p.y], [img_w, img_h]).astype(int) for p in results.multi_face_landmarks[0].landmark])
        except Exception as e:
            print(e)
            continue

            
        face_top = np.array((mesh_points[10][0],mesh_points[10][1]))
        face_bottom = np.array((mesh_points[152][0],mesh_points[152][1]))
        
        a = np.array((mesh_points[386][0],mesh_points[386][1]))
        b = np.array((mesh_points[374][0],mesh_points[374][1]))
        dist = np.linalg.norm(a-b)
        
        dist_face = np.linalg.norm(face_bottom-face_top)
        
        #print(dist, dist_face)
        #break
        
        if dist_face > 0:
            print(dist)
            #close_face.append(frame_name)
            if dist > 14.5:
                print("Alert")
                shutil.move(frame_path, "../Dataset/" + folder + "/Alert/")
            else:
                print("Drowsy")
                shutil.move(frame_path, "../Dataset/" + folder + "/Drowsy/")
        else:
            continue
        """
        
        cv2.polylines(frame_1, [mesh_points[LEFT_EYE]], True, (0,255,0), 1, cv2.LINE_AA)
        cv2.polylines(frame_1, [mesh_points[RIGHT_EYE]], True, (0,255,0), 1, cv2.LINE_AA)

        cv2.polylines(frame_1, [mesh_points[LEFT_IRIS]], True, (255,0,0), 1, cv2.LINE_AA)
        cv2.polylines(frame_1, [mesh_points[RIGHT_IRIS]], True, (255,0,0), 1, cv2.LINE_AA)

        cv2.polylines(frame_1, [mesh_points[MOUTH]], True, (0,255,0), 1, cv2.LINE_AA)

        cv2.imshow("frame", frame_1)
        cv2.waitKey(0)
        """
        
```
