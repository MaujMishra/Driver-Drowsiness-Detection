```python
import os
import random
import mediapipe as mp
import shutil
import cv2
import time
from PIL import Image 
import numpy as np

mp_face_mesh = mp.solutions.face_mesh
```

```python
list_dir = ['Fold1_part1_01_0',
 'Fold1_part1_01_1',
 'Fold1_part1_01_2',
 'Fold1_part1_02_0',
 'Fold1_part1_02_10',
 'Fold1_part1_02_5',
 'Fold1_part1_03_0',
 'Fold1_part1_03_10',
 'Fold1_part1_03_5',
 'Fold1_part1_05_0',
 'Fold1_part1_05_10',
 'Fold1_part1_05_5']
```

```python
dest = "../Dataset/Eye Dataset"
count_drowsy = 0
count_alert = 0
for i in list_dir:
    print(i)
    drowsy_path = os.path.join("../Dataset", i, "Drowsy")
    alert_path = os.path.join("../Dataset", i, "Alert")
    
    for j in os.listdir(drowsy_path):
        shutil.copy(os.path.join(drowsy_path, j), os.path.join(dest, "Drowsy"))
        os.rename(os.path.join(dest, "Drowsy", j), os.path.join(dest, "Drowsy", str(count_drowsy) + "_drowsy.jpg"))
        count_drowsy += 1
        
    for k in os.listdir(alert_path):
        shutil.copy(os.path.join(alert_path, k), os.path.join(dest, "Alert"))
        os.rename(os.path.join(dest, "Alert", k), os.path.join(dest, "Alert", str(count_alert) + "_alert.jpg"))
        count_alert += 1
    
```
