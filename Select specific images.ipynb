```python
import os
import shutil
import random
```

```python
alert_lst = os.listdir("../Dataset/Only Eyes Dataset/Alert")
drowsy_lst = os.listdir("../Dataset/Only Eyes Dataset/Drowsy")
```

```python
alert_lst[0: 5]
```

```python
grp_elems = {}

for elem in alert_lst:
    elem_split = elem.split("_")
    
    if elem_split[0] == "left" and  elem_split[1][0] != "f":
        try:
            grp_elems["left_only"] = grp_elems["left_only"] + [elem]
        except:
            grp_elems["left_only"] = [elem]
            
        continue
            
    if elem_split[0] == "right" and  elem_split[1][0] != "f":
        try:
            grp_elems["right_only"] = grp_elems["right_only"] + [elem]
        except:
            grp_elems["right_only"] = [elem]
        
        continue
    
    dict_key = elem_split[0] + "_" + elem_split[1] + "_" + elem_split[2] + "_" + elem_split[3]
    
    try:
        grp_elems[dict_key] = grp_elems[dict_key] + [elem]
    except:
        grp_elems[dict_key] = [elem]
    
    
    
```

```python
# get random 80 values from a list

for key in grp_elems.keys():
    value = grp_elems[key]
    selected_ix = []
    
    while (len(selected_ix) != 81):
        rand_ix = random.randint(0, len(value) -1)
        
        if rand_ix in selected_ix:
            continue
        else:
            selected_ix.append(rand_ix)
            selected_elem = value[rand_ix] 
            shutil.copy("../Dataset/Only Eyes Dataset/Alert/" + selected_elem, "../Dataset/Only Eyes Dataset/Updated Alert")

```
